```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
%matplotlib inline
```

```python
#carga de im√°genes y paso a RGB

tito_1 = cv.imread('./material_TPs/TP1/CoordCrom_1.png')
tito_1_RGB = cv.cvtColor(tito_1, cv.COLOR_BGR2RGB)

tito_2 = cv.imread('./material_TPs/TP1/CoordCrom_2.png')
tito_2_RGB = cv.cvtColor(tito_2, cv.COLOR_BGR2RGB)

tito_3 = cv.imread('./material_TPs/TP1/CoordCrom_3.png')
tito_3_RGB = cv.cvtColor(tito_3, cv.COLOR_BGR2RGB)
```

```python
plt.figure(1)
plt.imshow(tito_1_RGB)
plt.show()
print(tito_1_RGB.shape)
```

```python
plt.figure(1)
plt.imshow(tito_2_RGB)
plt.show()
print(tito_2_RGB.shape)
```

```python
plt.figure(1)
plt.imshow(tito_3_RGB)
plt.show()
print(tito_3_RGB.shape)
```

**TITO 1**

```python
#pasaje a coordenadas crom√°ticas

suma = tito_1_RGB.sum(axis = 2)

suma[suma==0]=0.1
        
tito1R = tito_1_RGB[:,:,0] / suma
tito1G = tito_1_RGB[:,:,1] / suma
tito1B = tito_1_RGB[:,:,2] / suma
```

```python
tito_1_RGB2 = cv.merge((np.uint8(tito1R*255), np.uint8(tito1G*255), np.uint8(tito1B*255)))
```

```python
plt.imshow(tito_1_RGB2)
plt.show()
tito_1_RGB2.dtype
```
