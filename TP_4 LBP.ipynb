```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from matplotlib import gridspec
```

```python
%matplotlib inline
```

```python
def get_pixel(img,x,y,center):
    
    new_value = 0
    
    try:
        if img[x][y] >= center:
            new_value = 1
        else:
            new_value = 0
    except:
        pass
    
    return new_value
```

```python
def lbp_calculate(img,x,y):
    
    center = img[x][y]
    
    pos = []
    
    pos.append(get_pixel(img, center, x-1, y+1))     
    pos.append(get_pixel(img, center, x, y+1))       
    pos.append(get_pixel(img, center, x+1, y+1))     
    pos.append(get_pixel(img, center, x+1, y))       
    pos.append(get_pixel(img, center, x+1, y-1))     
    pos.append(get_pixel(img, center, x, y-1))       
    pos.append(get_pixel(img, center, x-1, y-1))     
    pos.append(get_pixel(img, center, x-1, y))
    
    power_pos =[1,2,4,8,16,32,64,128]
    
    lum = 0
    
    for i in range(len(pos)):
        lum += pos[i] * power_pos[i]
    return lum
```

## Imagen 1

```python
img = cv.imread('./material_TPs/TP4/imagen1.png')
gray1 = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
gray1.shape
```

```python
plt.figure()
plt.imshow(gray1,cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
height, width = gray1.shape

img_lbp1 = np.zeros((height, width,3), np.uint8)

for i in range(0, height):
    for j in range(0, width):
        img_lbp1[i, j] = lbp_calculate(gray1, i, j)
hist_lbp1 = cv.calcHist([img_lbp1], [0], None, [256], [0, 256])


plt.imshow(img_lbp1)
```
