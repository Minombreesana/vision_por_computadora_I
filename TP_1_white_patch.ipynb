```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
%matplotlib inline
```

```python
#carga de im√°genes y paso a RGB

tito_1 = cv.imread('./material_TPs/TP1/WP_B.png')
tito_1_RGB = cv.cvtColor(tito_1, cv.COLOR_BGR2RGB)

tito_2 = cv.imread('./material_TPs/TP1/WP_O.png')
tito_2_RGB = cv.cvtColor(tito_2, cv.COLOR_BGR2RGB)

tito_3 = cv.imread('./material_TPs/TP1/WP_R.png')
tito_3_RGB = cv.cvtColor(tito_3, cv.COLOR_BGR2RGB)
```

```python
plt.figure(1)
plt.imshow(tito_1_RGB)
plt.show()
print(tito_1_RGB.shape)
```

```python
plt.figure(1)
plt.imshow(tito_2_RGB)
plt.show()
print(tito_2_RGB.shape)
```

```python
plt.figure(1)
plt.imshow(tito_3_RGB)
plt.show()
print(tito_3_RGB.shape)
```

**TITO AZUL**

```python
# Valores m√°ximos para cada canal

r_max = np.max(tito_1_RGB[:,:,0])
g_max = np.max(tito_1_RGB[:,:,1])
b_max = np.max(tito_1_RGB[:,:,2])

print('R_max:', r_max)
print('G_max:', g_max)
print('B_max:', b_max)
```

```python
# Normalizaci√≥n
 
tito_r = tito_1_RGB[:,:,0]/r_max
tito_g = tito_1_RGB[:,:,1]/g_max
tito_b = tito_1_RGB[:,:,2]/b_max
```

```python
tito_1_RGB2 = cv.merge((np.uint8(tito_r*255), np.uint8(tito_g*255), np.uint8(tito_b*255)))
```

```python
plt.imshow(tito_1_RGB2)
plt.show()
tito_1_RGB2.dtype
```

```python
# Prueba con percentiles

per = 98

r_per = np.percentile(tito_1_RGB[:,:,0], per)
g_per = np.percentile(tito_1_RGB[:,:,1], per)
b_per = np.percentile(tito_1_RGB[:,:,2], per)

print('R_per:', r_per)
print('G_per:', g_per)
print('B_per:', b_per)
```

```python
# Normalizaci√≥n
 
tito_r_per = tito_1_RGB[:,:,0]/r_per
tito_g_per = tito_1_RGB[:,:,1]/g_per
tito_b_per = tito_1_RGB[:,:,2]/b_per
```

```python
tito_1_RGB_per = cv.merge((np.uint8(tito_r_per*255), np.uint8(tito_g_per*255), np.uint8(tito_b_per*255)))
```

```python
plt.imshow(tito_1_RGB_per)
plt.show()
tito_1_RGB_per.dtype
```
